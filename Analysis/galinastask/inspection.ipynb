```python
import INS_Analysis as insd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from scipy.ndimage import gaussian_filter
```

```python
mctal_file = '../../simulations/galinastask/compute/output/mctal/Mo15Xfast.mctal'
```

```python
spectrums_file = '../../simulations/galinastask/data/spectrums.npz'
spectrums = np.load(spectrums_file)
spectrograms_file = '../../simulations/galinastask/data/spectrograms.npz'
spectrograms = np.load(spectrograms_file)
gebless_spectrums_file = '../../simulations/galinastask/data/gebless_spectrums.npz'
gebless_spectrums = np.load(gebless_spectrums_file)

bins = spectrums['x']
spectrums = spectrums['y'][:, 0, :]
spectrograms = spectrograms['y'][:, :, :]
gebless_spectrums = gebless_spectrums['y'][:, 0, :]
```

```python
plt.figure()
plt.plot(bins, spectrums[0])
plt.yscale('log')
plt.show()
```

```python
plt.figure()
plt.plot(bins, gebless_spectrums[0])
plt.yscale('log')
plt.show()
```

```python
spec = np.transpose(spectrograms[0])

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
plt.title('Spectrogram of trial 0001% C and 4666% Si (1e7 nps)')
fig.colorbar(ax.matshow(spec, aspect='auto', norm = plt.cm.colors.LogNorm()))
plt.ylim(200, 0)
plt.ylabel('Channel')
plt.xlabel('Time (ns)')
plt.show()
```

```python
spec = gaussian_filter(np.transpose(spectrograms[0]), sigma=(0, 1))
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
plt.title('Same spectrogram with a GEB for time blur (sigma = 1)')
fig.colorbar(ax.matshow(spec, aspect='auto', norm = plt.cm.colors.LogNorm()))
plt.ylim(200, 0)
plt.ylabel('Channel')
plt.xlabel('Time')
plt.show()
```
