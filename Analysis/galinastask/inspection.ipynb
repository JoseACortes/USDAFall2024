```python
mctal_file = '../../simulations/galinastask/compute/output/mctal/Mo15Xfast.mctal'
```

```python
spectrums_file = '../../simulations/galinastask/data/spectrums.npz'
spectrums = np.load(spectrums_file)
spectrograms_file = '../../simulations/galinastask/data/spectrograms.npz'
spectrograms = np.load(spectrograms_file)
gebless_spectrums_file = '../../simulations/galinastask/data/gebless_spectrums.npz'
gebless_spectrums = np.load(gebless_spectrums_file)

bins = spectrums['x']
spectrums = spectrums['y'][:, 0, :]
spectrograms = spectrograms['y'][:, :, :]
gebless_spectrums = gebless_spectrums['y'][:, 0, :]
```

```python
plt.figure(figsize=(3, 3), dpi=300)
plt.plot(bins, spectrums[0])
plt.yscale('log')
plt.title('INS Spectrum')
plt.xlabel('Energy [MeV]')
plt.ylabel('Counts')
plt.ylim(.5e0, 5e2)
plt.show()
```
